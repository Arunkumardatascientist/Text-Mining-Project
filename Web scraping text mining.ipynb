```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link='https://www.amazon.in/Renewed-OnePlus-Mirror-Black-Storage/product-reviews/B07KLG3KV1/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page=requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names=soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
len(names)
```

```python
cust_name=[]
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
    
```

```python
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(0)
```

```python
cust_name
```

```python
title=soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title=[]
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:]=[titles.rstrip('\n') for titles in review_title]
review_title  #lstrip and rstrip
```

```python
rating=soup.find_all('i',class_='review-rating')
```

```python
rating
```

```python
rate=[]
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
len(rate)
```

```python
rate.pop(0)
```

```python
rate.pop(0)
```

```python
review=soup.find_all('span',{'data-hook':'review-body'})
review
```

```python
review_content=[]
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:]=[reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
review_content[:]=[reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
cust_name
review_title
rate
review_content
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df
```

```python
df['customer_name']=cust_name
df['review_title']=review_title
df['Reviews']=review_content
df['Ratings']=rate
```
